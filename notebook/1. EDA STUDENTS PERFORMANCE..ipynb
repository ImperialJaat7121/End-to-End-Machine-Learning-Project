#### Exploring Data

```python
print("categories in 'gender' variable:    \n",end=" ")
print(df['gender'].unique())

print("categories in 'race/ethnicity' variable:    \n",end=" ")
print(df['race/ethnicity'].unique())

print("categories in 'parental level of education' variable:    \n",end=" ")
print(df['parental level of education'].unique())

print("categories in 'lunch' variable:    \n",end=" ")
print(df['lunch'].unique())

print("categories in 'test preparation course' variable:    \n",end=" ")
print(df['test preparation course'].unique())
```

define numerical and categorical features

```python
numerical_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']
```

print the columns

```python
print("We have {} numerical features and {} categorical features.".format(len(numerical_features), len(categorical_features)))
```

```python
df.head()
```

Adding columns for total score and average

```python
df["total score"] = df["math score"] + df["reading score"] + df["writing score"]
df["average score"] = df["total score"] / 3
```

```python
reading_full = df[df['reading score'] == 100]['average score'].count()
writing_full = df[df['writing score'] == 100]['average score'].count()
math_full = df[df['math score'] == 100]['average score'].count()

print(f"students with full reading score : {reading_full}")
print(f"students with full writing score : {writing_full}")
print(f"students with full math score : {math_full}")
```

```python
reading_less_20 = df[df['reading score'] <= 20]['average score'].count()
writing_less_20 = df[df['writing score'] <= 20]['average score'].count()
math_less_20 = df[df['math score'] <= 20]['average score'].count()

print(f"students with less than 20 reading score : {reading_less_20}")
print(f"students with less than 20 writing score : {writing_less_20}")
print(f"students with less than 20 math score : {math_less_20}")
```

#### Insights
- From above insights we get students have performed worst in Maths
- Best performance is in reading section

##### BIVARIATE ANALYSIS ( Is gender has any impact on student's performance ? )

```python
gender_group = df.groupby('gender')[numerical_features + ['total score', 'average score']].mean()
gender_group
```

```python
plt.figure(figsize=(10, 8))

X = ['Total Average','Math Average']


female_scores = [gender_group['average score'][0], gender_group['math score'][0]]
male_scores = [gender_group['average score'][1], gender_group['math score'][1]]

X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Female')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')
plt.legend()
plt.show()
```

#### Insights
- On an average females have a better overall score than men.
- whereas males have scored higher in Maths.
